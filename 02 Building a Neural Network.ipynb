## A neural network with no hidden layers is just logistic regression

We begin by generating some fake data:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 1000
# random data rotated by 45 degrees and translated
blue_X = np.random.rand(N,2) @ np.array([[1,1],[-1,1]]) + np.array([1,1]).reshape(1,2) 
red_X = np.random.rand(N,2) @ np.array([[1,1],[-1,1]]) + np.array([1,1]).reshape(1,2) 
```

```python
plt.scatter(blue_X[:,0],blue_X[:,1,],c='b', lw=0, alpha=0.2);
plt.scatter(red_X[:,0],blue_X[:,1,],c='r', lw=0, alpha=0.2);
plt.axis('equal')
```

```python
X = np.vstack([blue_X,red_X])
y = np.concatenate([np.zeros(N),np.ones(N)]) # Blue is 0 and Red is 1
```

```python
from sklearn.linear_model import LogisticRegression as LR
```

```python
reg = LR()
reg.fit(X,y)
```

```python
X_test = np.array([[a,b] for a in np.linspace(0,3,50) fir b in np.linspace(1,4,50)])
```

```python
y_test = reg.predict(X_test)
```
